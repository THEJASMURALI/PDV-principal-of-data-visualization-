```
x=[df[df['pdv score']<=40]['gender'].count(),df[df['pdv score']>=40]['gender'].count()]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
r = ['failed','passed']
ax.pie(x, labels = r,autopct='%1.2f%%')
plt.title("Pass and Fail percentage of pdv score")
plt.show()
```

```
x=[df[df['fml score']<=40]['gender'].count(),df[df['fml score']>=40]['gender'].count()]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
r = ['failed','passed']
ax.pie(x, labels = r,autopct='%1.2f%%')
plt.title("Pass and Fail percentage of fmlscore")
plt.show()
```

```
x=[df[df['pdv score']<=40]['gender'].count(),df[df['fml score']<=40]['gender'].count(),df[df['python score']<=40]['gender'].count()]
label=['PDV',"FML","PYTHON"]
plt.bar(label,x,color='r')
plt.xlabel("Subject")
plt.ylabel("Count")
plt.title("Number of students failed in each subject")
```

```
x=[df[df['pdv score']>=40]['gender'].count(),df[df['fml score']>=40]['gender'].count(),df[df['python score']>=40]['gender'].count()]
label=['PDV',"FML","PYTHON"]
plt.bar(label,x,color='g')
plt.xlabel("Subject")
plt.ylabel("Count")
plt.title("Number of students failed in each subject")
```

```
pass_c=df[(df['fml score']>=40) & (df['pdv score']>=40) & (df['python score']>=40)]['gender'].count()
fail_c=1000-pass_c
```

```
plt.bar(['Passed','Failed'],[pass_c,fail_c],color='b')
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
r = ['passed','failed']
ax.pie([pass_c,fail_c], labels = r,autopct='%1.2f%%')
plt.title("Pass and Fail percentage of students")
plt.show()
```
