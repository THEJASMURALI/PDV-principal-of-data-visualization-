```
# Load EDA Packages
import pandas as pd
import numpy as np     # To work with Arrays and neumarical analysis.
import matplotlib.pyplot as plt  # For visualization.
import seaborn as sns  # For visualization and to apply statistical functions.
# For displaying graphs within jupyter notebook.
%matplotlib inline 
```

```
df = pd.read_csv('items.csv')
```

```
df.head()
```

```
df['price'] = df['price'].str.replace(',', '') 
df
```

```
df['ratings'] = df['ratings'].str.replace(',', '')
df
```

```
df["model_name"] = df["model_name"].str.split(" ", n = 1, expand = True)[0]
df
```

```
df.describe()
```

```
# Missing Data 
df.isnull()
```

```
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

```
sns.set_style('whitegrid')
sns.countplot(x='model_name',data=df)
```

```
sns.set_style('whitegrid')
sns.countplot(x='model_name',hue='price',data=df,palette='RdBu_r')
```

```
sns.set_style('whitegrid')
sns.countplot(x='model_name',hue='price',data=df,palette='rainbow')
```
